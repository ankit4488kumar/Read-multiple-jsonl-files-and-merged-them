```python
import json
```

```python
with open('validation_1.jsonl', 'r') as json_file1:
    json_list1 = list(json_file1)

for json_str1 in json_list1:
    result1 = json.loads(json_str1)
    print(f"result1: {result1}")
    print(isinstance(result1, dict))
```

```python
with open('train_1.jsonl', 'rb') as json_file2:
    json_list2 = list(json_file2)

for json_str2 in json_list2:
    result2 = json.loads(json_str2)
    print(f"result2: {result2}")
    print(isinstance(result2, dict))
```
